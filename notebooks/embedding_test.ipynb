```python
import pickle
```

# Load Chunks

```python
with open("../data/chunks.pickle", "rb") as f:
    chunks = pickle.load(f)
```

```python
print(chunks[3].page_content)
```

# Load Embeddings Model

```python
from langchain_ollama import OllamaEmbeddings
```

```python
EMBEDDING_MODEL = 'hf.co/Casual-Autopsy/snowflake-arctic-embed-l-v2.0-gguf:Q4_K_S'
OLLAMA_EMBEDDING_BASE_URL = 'http://127.0.0.1:11434'

embeddings = OllamaEmbeddings(
    model=EMBEDDING_MODEL,
    base_url=OLLAMA_EMBEDDING_BASE_URL,
)

embeddings
```

# Create Vetors Data Base Store

```python
from langchain_community.vectorstores import Chroma
```

```python
persist_directory = "./chroma_db"

vectorstore = Chroma.from_documents(
    documents=chunks,
    embedding=embeddings,
    persist_directory=persist_directory,
)

vectorstore
```

# Retriever

```python
retriever = vectorstore.as_retriever(search_kwargs={"k": 5})
retriever
```
