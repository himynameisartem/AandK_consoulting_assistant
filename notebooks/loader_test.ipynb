```python
import os
import nest_asyncio
from  langchain_community.document_loaders import SitemapLoader
```

```python
os.environ["USER_AGENT"] = "Mozilla/5.0"
nest_asyncio.apply()
```

# Download

```python
ROOT_URL = "https://us-ak.com/"
SITEMAP_URL = f"{ROOT_URL}sitemap.xml"
```

```python
sitemap_loader = SitemapLoader(
    web_path=SITEMAP_URL,
    filter_urls=[ROOT_URL]
)
docs = sitemap_loader.load()
sitemap_docs = docs
```

```python
print(f"Found {len(docs)} documents")
print(f"Total characters: {sum(len(doc.page_content) for doc in docs)}")
```

# Text Splittters

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter, Language
```

```python
text_splitter = RecursiveCharacterTextSplitter.from_language(
    language=Language.HTML,
    chunk_size=500,
    chunk_overlap=100
)
splits = text_splitter.split_documents(docs)
```

```python
print(splits[0].page_content.replace('\n', ' '))
print("---------------------------------------")
print(splits[1].page_content.replace('\n', ' '))
print("---------------------------------------")
print(splits[2].page_content.replace('\n', ' '))
```

# Save Chunks

```python
with open ("../data/chunks.pickle", "wb") as f:
    pickle.dump(splits, f)
```
